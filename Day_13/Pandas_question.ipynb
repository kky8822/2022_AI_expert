E1. 아래의 코드를 실행해 보면서 series와 dataframe을 만드는 법을 알아봅시다. (**제한시간 5분**)

```python
import pandas as pd
import numpy as np

datalist = list('abcdefg')
dataarr = np.arange(7)
datadict = dict(zip(datalist, dataarr))

ser1 = pd.Series(datalist)
ser2 = pd.Series(dataarr)
ser3 = pd.Series(datadict)
print(ser3)

df = ser3.to_frame().reset_index()
print(df)

df = pd.DataFrame({'col1':ser1, 'col2':ser2})
print(df)
```
