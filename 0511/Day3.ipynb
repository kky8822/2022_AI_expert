# Factorial
팩토리얼 n! 은 1부터 n까지의 정수의 곱이다.
recursion을 이용하여 n!을 구하는 함수를 만들어 보자.

## 내가 푼거

```python
def fact(n):
    if n == 1:
        return 1
    return fact(n-1) * n

print(fact(4))
```

## 풀이

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)  # recursion


def factorial2(n):
    value = 1
    for i in range(1, n + 1):
        value *= i
    return value


my_n = input("Factorial of: ")
my_n = int(my_n)
print("Result: ", factorial(my_n), factorial2(my_n))
```

# 도서 대출 프로그램
Dictionary 와 class를 이용하여 잔여 도서를 count하고 대출하는 모듈을 구성한다. 구현해야 할 기능은 이러하다. 대출자의 이름에 중복은 없다고 가정한다.
- 클래스를 만들어 멤버 변수와 메소드로 도서를 관리한다. 
- 각 도서별로 이름과 수량을 dictionary로 구현한다.
- 이름을 input으로 받아 도서를 색인하는 함수를 만든다. 
- 대출자의 이름과 책의 이름을 받아 책을 대출하는 함수를 구현한다. 대출자가 이미 대출하고 있는 책이 없고, 책의 수량이 0이 아니라면 책을 대출할 수 있다.
- 대출자의 이름을 받아 대출자가 가진 책을 반납하는 함수를 구현한다.


## 내가 푼거

```python
class library:
    members = {"kky":None, "yjh":None, "awesome":None}
    books = {"a":10, "b":2, "c":1}

    @staticmethod
    def indexing_book():
        book_name = input("What's name of book ?: ")
        q = library.books[book_name]
        print(f"book name:{book_name}, q: {q}")

    @staticmethod
    def check_out(member, book_name):
        if library.members[member] == None and library.books[book_name] != 0:
            library.set_members(member, book_name, "check_out")
            library.set_books(book_name, "check_out")
            return "Success"
        else:
            return "Fail"

    @staticmethod
    def check_in(member):
        if library.members[member] != None:
            book_name = library.members[member]
            library.set_members(member, book_name, "check_in")
            library.set_books(book_name, "check_in")
            return "Success"
        else:
            return "Fail"

    @staticmethod
    def set_members(member, book_name, form):
        if form == "check_out":
            library.members[member] = book_name
        elif form == "check_in":
            library.members[member] = None

    @staticmethod
    def set_books(book_name, form):
        if form == "check_out":
            library.books[book_name] = library.books[book_name]-1
        elif form == "check_in":
            library.books[book_name] = library.books[book_name]+1


# library.indexing_book()

print(library.members)
print(library.books)

library.check_out("kky", "a")
print(library.members)
print(library.books)

library.check_out("yjh", "c")
print(library.members)
print(library.books)

library.check_out("awesome", "c")
print(library.members)
print(library.books)

library.check_in("kky")
print(library.members)
print(library.books)
```

## 풀이

```python
class Library():

    def __init__(self):
        self.book = {"batman": 3, "superman": 3, "antman": 3}
        self.state = dict()  # who : what

    def add(self, book_name, book_num=1):
        if book_name not in self.book:
            self.book[book_name] = book_num
        else:
            self.book[book_name] += book_num
        self.find(book_name)

    def find(self, book_name):
        if book_name not in self.book:
            print("Book does not exist")
            return

        print("Book {} left: {}".format(book_name, self.book[book_name]))

    def borrow(self, person_name, book_name):
        person_name = str(person_name)
        if person_name in self.state:
            print("Already borrowed book.")
        elif book_name not in self.book:
            print("Book does not exist.")
        elif self.book[book_name] < 1:
            print("No book left.")
        else:
            print("Borrowing book")
            self.book[book_name] -= 1
            self.state[person_name] = book_name

    def back(self, person_name):
        person_name = str(person_name)
        if person_name not in self.state:
            print("Nobody of that name.")
        else:
            borrowed_book = self.state[person_name]
            print("Book {} returned.".format(borrowed_book))
            self.book[borrowed_book] += 1
            del self.state[person_name]


lib = Library()
lib.find("antman")  # Book antman left: 3
lib.find("ironman")  # Book does not exist
lib.add("ironman", 2)  # Book ironman left: 2

lib.borrow("Choi", "ironman")  # Borrowing book
lib.find("ironman")  # Book ironman left: 1

lib.borrow("Kim", "aquaman")  # Book does not exist
lib.borrow("Kim", "ironman")  # Borrowing book
lib.borrow("Kim", "antman")  # Already borrowed book

lib.borrow("Lee", "ironman")  # No book left
lib.back("Kim")  # Book ironman returned
lib.borrow("Lee", "ironman")  # Borrowing book
```

# 괴물 사냥 게임
3가지 종류의 괴물 (거미, 뱀, 박쥐)이 있다.
- 각각의 체력(HP)은 5, 10, 20 이고, 이 HP가 다 사라지면 죽는다.
- 각각의 공격력(ATK)은 1, 2, 3 이다.
- 턴이 1 지날 때마다 체력이 1 증가한다.
- 사용자는 HP 25, 공격력은 4이다.
- 턴이 1 지날 때마다 공격력이 1 증가한다.

괴물 6마리 (거미x3, 뱀x2, 박쥐x1) 을 상대로
사용자 1번 – 괴물(순서대로) 1번 - …  순서로 공격한다.
사용자는 매 턴 공격할 괴물을 입력한다.
괴물을 다 잡아 보자.


## 내가 푼거

```python
class Game:
    def __init__(self):
        self.monsters = [
            {"name":"spider1", "HP":5, "ATK":1},
            {"name":"spider2", "HP":5, "ATK":1}, 
            {"name":"spider3", "HP":5, "ATK":1},         
            {"name":"snake1", "HP":10, "ATK":2}, 
            {"name":"snake2", "HP":10, "ATK":2},        
            {"name":"bat1", "HP":20, "ATK":3}, 
        ]
        self.user = {"HP":25, "ATK":4}
        self.state()

    def state(self):
        print(f"user HP: {self.user['HP']}, ATK: {self.user['ATK']}")
        for i, monster in enumerate(self.monsters):
            print(f"{monster['name']} HP: {monster['HP']}, ATK: {monster['ATK']}")
        print("-----------------------------")

    def die(self):
        for i, monster in enumerate(self.monsters):
            if self.monsters[i]["HP"] <= 0:
                del self.monsters[i]

    def run(self):
        count = 0
        while self.user["HP"] > 0 and len(self.monsters) > 0:
            
            target=input("Target :")
            for i, monster in enumerate(self.monsters):
                if monster["name"] == target:
                    self.monsters[i]["HP"] = self.monsters[i]["HP"] - self.user["ATK"]
            self.die()

            if len(self.monsters) < 1:
                print("You Win!")
                break
    
            self.user["HP"] = self.user["HP"] - self.monsters[count]["ATK"]

            if self.user["HP"] <= 0:
                print("You lose")
                break
            
            self.user["ATK"] = self.user["ATK"] + 1
            for i, monster in enumerate(self.monsters):
                self.monsters[i]["HP"] = self.monsters[i]["HP"] + 1
            
            self.state()

go = Game()
go.run()


```

## 풀이

```python
class Living:

    def __init__(self, hp, atk, name):
        self.hp = hp
        self.atk = atk
        self.died = False
        self.name = name

    def hit(self, atk):
        if self.died:
            print("{} already died.".format(self.name))
            return

        self.hp = self.hp - atk
        print("{} got hit! HP remain: {}.".format(self.name, self.hp))
        if self.hp <= 0:
            print("{} died.".format(self.name))
            self.died = True

    def turn_finish(self):
        pass

class Monster(Living):

    def __init__(self, hp, atk, name):
        Living.__init__(self, hp, atk, name)

    def turn_finish(self):
        self.hp += 1


class Spider(Monster):

    def __init__(self, name):
        Monster.__init__(self, 5, 1, name)


class Snake(Monster):

    def __init__(self, name):
        Monster.__init__(self, 10, 2, name)


class Bat(Monster):

    def __init__(self, name):
        Monster.__init__(self, 20, 3, name)


class User(Living):

    def __init__(self):
        Living.__init__(self, 25, 4, 'User')

    def turn_finish(self):
        self.atk += 1

    def print_user(self):
        print("HP: {}, ATK: {}".format(self.hp, self.atk))
```

```python
# from day3_example3_utils import User, Spider, Snake, Bat

class Enemy:

    def __init__(self, monsters):
        self.monsters = monsters
        self._counter = 0

    def monster_names(self):
        return [mon.name for mon in self.monsters]

    def check_alive(self):
        living_monsters = []
        for mon in self.monsters:
            if not mon.died:
                living_monsters.append(mon)
        self.monsters = living_monsters
        return len(self.monsters) > 0

    def find_monster_by_name(self, monster_name):
        for mon in self.monsters:
            if mon.name == monster_name:
                return mon
        print("Monster does not exist")
        return None

    def turn_finish(self):
        for mon in self.monsters:
            mon.turn_finish()

    def get_atk(self):
        if self._counter >= len(self.monsters):
            self._counter = 0
            
        mon = self.monsters[self._counter]
        print("{} is attacking with {}".format(mon.name, mon.atk))
        self._counter += 1

        return mon.atk

    def print_monsters(self):
        print('-' * 20)
        for mon in self.monsters:
            print("{} (HP: {}, ATK: {})".format(mon.name, mon.hp, mon.atk))

user = User()
monsters = [Spider('sp1'), Spider('sp2'), Spider('sp3'), Snake('sn1'), Snake('sn2'), Bat('ba1')]
enemy = Enemy(monsters)

while True:
    enemy.print_monsters()
    user.print_user()

    # ---------------------------------------------------------------- #
    # User turn
    while True:
        what_to_attack = input("What to attack?: ")
        if what_to_attack in enemy.monster_names():
            break
        print("Monster does not exist.")

    target_monster = enemy.find_monster_by_name(what_to_attack)
    target_monster.hit(user.atk)

    # ---------------------------------------------------------------- #
    # Monster turn
    user.hit(enemy.get_atk())

    # ---------------------------------------------------------------- #
    # Check quit
    enemy_left = enemy.check_alive()
    if (not user.died) and enemy_left:
        user.turn_finish()
        enemy.turn_finish()
        continue
    elif not enemy_left:
        print("Victory!")
        break
    else:
        print("Defeat!")
        break
    
```
