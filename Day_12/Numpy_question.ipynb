Q3. 이번에는 Condition을 사용하여 짝수값들을 음수로 만들어 봅시다. (**제한시간 3분**)

```python
nums = np.arange(3,12)
nums[np.where(nums%2 ==0)] = nums[nums%2==0]*-1

print(nums)
```

E2. Conditional indexing을 사용하여 element의 값을 바꿀 경우, 원본 array의 값이 바뀌는 문제가 있습니다. 

```python
nums = np.arange(3,12)
nums[np.where(nums%2 ==0)] = nums[nums%2==0]*-1

print(nums)
```

E2 & Q4. 원본 array에 영향을 주지 않으면서 Condition을 통해 element를 선택하고 싶을때는 where을 사용하면 됩니다. 한번 Where를 사용해봅시다. (**제한시간 8분**)

```python
nums = np.arange(3, 12)
out = np.where(nums%2==0)

print(nums)
print(out)
```
